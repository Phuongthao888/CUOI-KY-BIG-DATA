```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lay du lieu 
path = '/home/phuongthao/bigdata/product.csv'
dataframe = pd.read_csv(path)
dataframe.head(10)
```

```python
#Kiem tra kieu du lieu cua cot Date
print(dataframe['Date'].dtype)
```

```python
#them cot month, de thong ke so lieu de dang hon
dataframe['Month'] = ''
```

```python
#add thang vua tach ra tu cot Date vao cot Month
dataframe['Month'] = dataframe['Date'].str.slice(0,2,1)
dataframe.head(10)
```

```python
#kiem tra lai cot thang bang cach in ra cac thang co trong cot
print(set(dataframe['Month']))
```

```python
#Ta doi cot Month len phia truoc cho de quan sat
column = dataframe.pop('Month')
dataframe.insert(1,'Month', column)
dataframe.head(10)
```

```python
#Câu hỏi số 1: Loại sản phẩm nào được khách hàng quan tâm Reaction nhiều nhất trong năm? 
```

```python
#tinh tong luong Reaction cua cac Category Product Facebook
dataframe.groupby('Category').sum()['Reaction']
```

```python
#Luong reaction cao nhat cua mot Category Product Facebook
reaction_max = dataframe.groupby('Category').sum()['Reaction']
reaction_max.max()
```

```python
#Lay tat ca Category ProductFB gan vao 1 bien
nameCa = dataframe['Category'].unique()
```

```python
plt.bar(x=nameCa, height=reaction_max) 
plt.xticks(nameCa, rotation = 90) #Hien ten Category o cot gia tri, chinh chu nghieng 
plt.xlabel('Category') #ten hang ngang
plt.ylabel('Category') #ten chieu doc
plt.title('Số lượng Category cao nhất cho một sản phẩm')
plt.show()
```

```python
#Câu hỏi số 2: Facebook Product Category nào được người dùng comment nhiều nhất?
```

```python
#Groupby cac Category cung loai, hien ra tong luot Comments cua Category Product Facebook trong 1 nam 
dataframe.groupby('Category').sum()['Comments']
```

```python
#Hien ra luong Comments cao nhat trong nam
cmt_max = dataframe.groupby('Category').sum()['Comments']
cmt_max.max()
```

```python
#Gan tat ca cac Catetogy vao 1 bien 
nameCmt = dataframe['Category'].unique()

plt.bar(x = nameCmt, height = cmt_max)
plt.xticks(nameCmt, rotation = 90)
plt.xlabel('Category')
plt.ylabel('Comments')
plt.title('Số lượng bình luận cao nhất cho một sản phẩm')
plt.show()
```

```python
#Câu hỏi số 3: có bao nhiêu sản phẩm trong 1 Facebook Product Category?
```

```python
#Dem luong san pham cua cac Category Product Facebook
dataframe.groupby('Category').count()['Product Name']
```

```python
#Sau khi dem so luong san pham cua 1 Category, ta gan vao 1 bien
countSP = dataframe.groupby('Category').count()['Product Name']
#Gan tat ca cac Catetogy vao 1 bien 
nameProName = dataframe['Category'].unique()
#Dung plt
plt.bar(x = nameProName, height = countSP)
plt.xticks(nameProName, rotation = 90)
plt.xlabel('Category')
plt.ylabel('So Luong Product Name')
plt.title('Sản phẩm được mọi người quan tâm nhất')
plt.show()
```

```python
#Câu hỏi số 4: Sản phẩm được độ tuổi nào quan tâm nhiều nhất?
```

```python
Clothing_pro = dataframe[dataframe['Category'] == 'Body Paint']
AgeCount = Clothing_pro['Age'].value_counts()
AgeCount = Clothing_pro['Age'].value_counts().sort_index()

# Vẽ biểu đồ đường
AgeCount.plot.line(marker='o', linestyle='-', color='red')

plt.title('Số lượng người sử dụng theo độ tuổi cho loại sản phẩm')
plt.xlabel('Độ Tuổi')
plt.ylabel('Số lượng người sử dụng')
plt.show()
```

```python
#Câu hỏi số 6: Tính tổng tỷ lệ của từng độ tuổi quan tâm đến Facebook Product Category
```

```python
Age_sort = dataframe.sort_values(by='Age')
Age_ratio = Age_sort['Age'].value_counts(normalize=True)

# Tạo biểu đồ pie chart tổng hợp
Age_ratio.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Overall Age Distribution')
plt.ylabel('Age_ratio')  
plt.show()
```
